# **Decision Tree for Classification**

## **Gini**

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

data = {'age': [23, 25, 27, 29, 29],
        'likes english': [0, 1, 1, 0, 0],
        'likes ai': [0, 1, 0, 1, 0],
        'raise salary': [0, 0, 1, 1, 0]}

# load data
df = pd.DataFrame(data)

# get value and convert to numpy
X = df[['age', 'likes english', 'likes ai']].values
y = df[['raise salary']].values.reshape(-1,)

# define classifier
clf = DecisionTreeClassifier()

# train
clf = clf.fit(X, y)
```

```python
X
```

```python
y
```

```python
# predict

x_test = np.array([[27, 0, 1]])
predicted_label = clf.predict(x_test)
predicted_label
```

```python
from sklearn.tree import plot_tree

# visualization
plot_tree(clf, feature_names=['age', 'likes english', 'likes ai'], fontsize=10)
```

## **Information Gain**

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier

data = {'age': [23, 25, 27, 29, 29],
        'likes english': [0, 1, 1, 0, 0],
        'likes ai': [0, 1, 0, 1, 0],
        'raise salary': [0, 0, 1, 1, 0]}

# load data
df = pd.DataFrame(data)

# get value and convert to numpy
X = df[['age', 'likes english', 'likes ai']].values
y = df[['raise salary']].values.reshape(-1,)

# define classifier
clf = DecisionTreeClassifier(criterion='entropy')

# train
clf = clf.fit(X, y)
```

```python
# predict
x_test = np.array([[27, 0, 1]])
predicted_label = clf.predict(x_test)
predicted_label
```

```python
from sklearn.tree import plot_tree

# visualization
plot_tree(clf, feature_names=['age', 'likes english', 'likes ai'], fontsize=10)
```

##**IRIS Dataset**

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.tree import DecisionTreeClassifier
```

```python
# Load the diabetes dataset
iris_X, iris_y = datasets.load_iris(return_X_y=True)
```

```python
iris_X[:5]
```

```python
iris_y[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_X, iris_y, test_size=0.2, random_state=7
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_pred, y_test)
```

```python
from sklearn.tree import plot_tree

# visualization
plot_tree(clf, feature_names=["Sepal_Length", "Sepal_Width", "Petal_Length", "Petal_Width"])
```

# **Decision Tree for Regreesion**

## **Sum of Squared Error**

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor

data = {'age': [23, 25, 27, 29, 29],
        'likes english': [0, 1, 1, 0, 0],
        'likes ai': [0, 1, 0, 1, 0],
        'salary': [200, 400, 300, 500, 400]}

# load data
df = pd.DataFrame(data)
```

```python
df
```

```python
# get value and convert to numpy
X = df[['age', 'likes english', 'likes ai']].values
y = df[['salary']].values.reshape(-1,)

# define regressor
reg = DecisionTreeRegressor()

# train
reg = reg.fit(X, y)
```

```python
# predict
x_test = np.array([[27, 0, 1]])
predicted_label = reg.predict(x_test)
predicted_label
```

```python
from sklearn.tree import plot_tree

# visualization
plot_tree(reg, feature_names=['age', 'likes english', 'likes ai'], fontsize=10)
```

## **CPU Machine Dataset**

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.datasets import fetch_openml

machine_cpu = fetch_openml(name='machine_cpu')
```

```python
print(machine_cpu.DESCR)
```

```python
x_data = machine_cpu.data
y_data = machine_cpu.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_data, y_data, test_size=0.2, random_state=7
)
```

```python
# define regressor
reg = DecisionTreeRegressor(max_depth=1)

# train
reg = reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
mean_squared_error(y_pred, y_test)
```

```python
from sklearn.tree import plot_tree

# visualization
plot_tree(reg, feature_names=['MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN', 'CHMAX'])
```
